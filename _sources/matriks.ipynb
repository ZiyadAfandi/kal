# Matriks (Penjumlahan, Pengurangan, dan Perkalian)

## 1. Pengertian Dasar Matriks
Matriks adalah susunan bilangan, simbol, atau ekspresi yang diatur dalam baris dan kolom sehingga membentuk persegi atau persegi panjang. Ukuran suatu matriks dinyatakan dalam **ordo**, yaitu $m \times n$, di mana $m$ melambangkan jumlah baris dan $n$ melambangkan jumlah kolom.

---

## 2. Penjumlahan Matriks

### Syarat Penjumlahan
Dua matriks atau lebih dapat dijumlahkan **hanya jika memiliki ordo (ukuran) yang sama**. Jika ordonya berbeda, maka operasi penjumlahan tidak dapat dilakukan.

### Cara Menghitung
Penjumlahan dilakukan dengan cara menjumlahkan elemen-elemen yang **seletak** (berada pada baris dan kolom yang sama).

### Bentuk Umum
Jika Matriks $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$ dan $B = \begin{pmatrix} e & f \\ g & h \end{pmatrix}$, maka:

$$A + B = \begin{pmatrix} a+e & b+f \\ c+g & d+h \end{pmatrix}$$

### Contoh Soal
Diketahui matriks:
$$A = \begin{pmatrix} 2 & 4 \\ 1 & 3 \end{pmatrix}, \quad B = \begin{pmatrix} 5 & 1 \\ 0 & 2 \end{pmatrix}$$

Tentukan hasil dari $A + B$!

**Penyelesaian:**
$$A + B = \begin{pmatrix} 2+5 & 4+1 \\ 1+0 & 3+2 \end{pmatrix} = \begin{pmatrix} 7 & 5 \\ 1 & 5 \end{pmatrix}$$

---

## 3. Pengurangan Matriks

### Syarat Pengurangan
Sama seperti penjumlahan, pengurangan matriks hanya bisa dilakukan jika **kedua matriks memiliki ordo yang sama**.

### Cara Menghitung
Pengurangan dilakukan dengan mengurangkan elemen-elemen yang **seletak** antara matriks pertama dengan matriks kedua.

### Bentuk Umum
Jika Matriks $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$ dan $B = \begin{pmatrix} e & f \\ g & h \end{pmatrix}$, maka:

$$A - B = \begin{pmatrix} a-e & b-f \\ c-g & d-h \end{pmatrix}$$

### Contoh Soal
Menggunakan matriks $A$ dan $B$ dari contoh sebelumnya, tentukan hasil dari $A - B$!

**Penyelesaian:**
$$A - B = \begin{pmatrix} 2-5 & 4-1 \\ 1-0 & 3-2 \end{pmatrix} = \begin{pmatrix} -3 & 3 \\ 1 & 1 \end{pmatrix}$$

---

## 4. Perkalian Matriks

Operasi perkalian pada matriks dibagi menjadi dua jenis, yaitu perkalian dengan skalar dan perkalian dua matriks.

### A. Perkalian Matriks dengan Skalar (Angka Real)
Perkalian skalar terjadi ketika sebuah matriks dikalikan dengan sebuah angka biasa (skalar). Caranya adalah dengan **mengalikan seluruh elemen di dalam matriks** dengan angka tersebut.

* **Bentuk Umum:**
  $$k \cdot \begin{pmatrix} a & b \\ c & d \end{pmatrix} = \begin{pmatrix} k \cdot a & k \cdot b \\ k \cdot c & k \cdot d \end{pmatrix}$$

* **Contoh Soal:**
  $$3 \cdot \begin{pmatrix} 2 & -1 \\ 4 & 0 \end{pmatrix} = \begin{pmatrix} 3(2) & 3(-1) \\ 3(4) & 3(0) \end{pmatrix} = \begin{pmatrix} 6 & -3 \\ 12 & 0 \end{pmatrix}$$

### B. Perkalian Dua Matriks
Dua buah matriks dapat dikalikan jika **jumlah kolom pada matriks pertama (kiri) sama dengan jumlah baris pada matriks kedua (kanan)**. 

* **Prinsip Dasar:** Mengalikan **Baris** pada matriks pertama dengan **Kolom** pada matriks kedua, kemudian menjumlahkan hasilnya.



* **Bentuk Umum ($2 \times 2$):**
  Jika $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$ dan $B = \begin{pmatrix} e & f \\ g & h \end{pmatrix}$, maka:
  $$A \times B = \begin{pmatrix} (a \cdot e + b \cdot g) & (a \cdot f + b \cdot h) \\ (c \cdot e + d \cdot g) & (c \cdot f + d \cdot h) \end{pmatrix}$$

* **Contoh Soal:**
  Diketahui $A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$ dan $B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}$. Tentukan $A \times B$!

  **Penyelesaian:**
  * Elemen baris 1, kolom 1: $(1 \cdot 5) + (2 \cdot 7) = 5 + 14 = 19$
  * Elemen baris 1, kolom 2: $(1 \cdot 6) + (2 \cdot 8) = 6 + 16 = 22$
  * Elemen baris 2, kolom 1: $(3 \cdot 5) + (4 \cdot 7) = 15 + 28 = 43$
  * Elemen baris 2, kolom 2: $(3 \cdot 6) + (4 \cdot 8) = 18 + 32 = 50$

  Sehingga diperoleh:
  $$A \times B = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}$$

> ⚠️ **Sifat Penting:** Perkalian dua matriks **tidak bersifat komutatif** ($A \times B \neq B \times A$). Jadi, urutan perkalian sangat menentukan hasil.

---

## 5. Ringkasan Aturan Ordo Matriks

| Jenis Operasi | Syarat Ordo Matriks | Ordo Matriks Hasil |
| :--- | :--- | :--- |
| **Penjumlahan ($A + B$)** | Ordo harus sama ($m \times n$ dan $m \times n$) | $m \times n$ |
| **Pengurangan ($A - B$)** | Ordo harus sama ($m \times n$ dan $m \times n$) | $m \times n$ |
| **Perkalian ($A \times B$)** | Kolom matriks $A$ = Baris matriks $B$ ($m \times \mathbf{n}$ dan $\mathbf{n} \times p$) | $m \times p$ |

---
*Tips Belajar: Untuk menghindari kekeliruan saat mengalikan dua matriks, buatlah garis penanda horizontal pada baris matriks pertama dan garis penanda vertikal pada kolom matriks kedua.*
