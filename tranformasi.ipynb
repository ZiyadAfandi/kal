## Transformasi Matriks


## Translasi adalah jenis transformasi geometri yang memindahkan setiap titik pada bidang dengan jarak dan arah yang tetap. Dalam bentuk matriks, translasi $T$ dinyatakan sebagai:
$$
T = \begin{pmatrix} 
a \\ b \end{pmatrix}
$$
Metodologi PerhitunganUntuk menemukan nilai matriks translasi dari titik asal 
$$
(x, y)
$$ 
ke titik tujuan 
$$(x', y')$$, 
menghitung selisih koordinatnya:
$a$ (Sumbu X): $x' - x$ (positif ke kanan, negatif ke kiri)
$b$ (Sumbu Y): $y' - y$ (positif ke atas, negatif ke bawah)
Hasil Transformasi Titik
# 1. Transformasi A ke BKoordinat: 
$$A(2, 3) \rightarrow B(2, 1)$$
Langkah:$$a = 2 - 2 = 0$$ $$b = 1 - 3 = -2$$
Matriks: $$\begin{pmatrix} 0 \\ -2 \end{pmatrix}$$
# 2. Transformasi B ke CKoordinat: 
$$B(2, 1) \rightarrow C(4, 1)$$
Langkah:$$a = 4 - 2 = 2$$ $$b = 1 - 1 = 0$$
Matriks: $$\begin{pmatrix} 2 \\ 0 \end{pmatrix}$$
# 3. Transformasi D ke EKoordinat: 
$$D(2, 4) \rightarrow E(2, 0)$$
Langkah:$$a = 2 - 2 = 0$$ $$b = 0 - 4 = -4$$
Matriks: $$\begin{pmatrix} 0 \\ -4 \end{pmatrix}$$
# 4. Transformasi E ke FKoordinat: 
$$E(2, 0) \rightarrow F(4, 0)$$
Langkah:$$a = 4 - 2 = 2$$ $$b = 0 - 0 = 0$$
Matriks: $$\begin{pmatrix} 2 \\ 0 \end{pmatrix}$$
