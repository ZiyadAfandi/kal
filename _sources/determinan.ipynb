# Determinan Matriks Ordo 2x2

## 1. Pengertian Determinan Matriks
Determinan adalah sebuah nilai skalar (angka real) yang dihasilkan dari elemen-elemen suatu matriks persegi. Determinan dari matriks $A$ biasanya dinotasikan dengan lambang **$\det(A)$** atau **$|A|$**.

*Catatan: Hanya matriks persegi (jumlah baris dan kolom sama) yang memiliki nilai determinan.*

---

## 2. Rumus Determinan Matriks Ordo 2x2

Matriks ordo $2 \times 2$ terdiri dari 2 baris dan 2 kolom (total 4 elemen). 

Jika kita memiliki matriks $A$ sebagai berikut:
$$A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$$

Maka, determinan dari matriks $A$ dihitung dengan cara **mengurangi hasil kali elemen-elemen diagonal utama dengan hasil kali elemen-elemen diagonal sekunder**.



### Rumus:
$$\det(A) = |A| = (a \cdot d) - (b \cdot c)$$

**Keterangan:**
* $a$ dan $d$ adalah elemen **diagonal utama** (dari kiri atas ke kanan bawah).
* $b$ oars $c$ adalah elemen **diagonal sekunder** (dari kanan atas ke kiri bawah).

---

## 3. Contoh Soal dan Penyelesaian

### Contoh 1 (Bilangan Positif)
Diketahui matriks $B = \begin{pmatrix} 5 & 3 \\ 2 & 4 \end{pmatrix}$. Tentukan determinan matriks $B$!

**Penyelesaian:**
* $a = 5, b = 3, c = 2, d = 4$
$$\det(B) = (5 \cdot 4) - (3 \cdot 2)$$
$$\det(B) = 20 - 6$$
$$\det(B) = 14$$
Jadi, nilai determinan matriks $B$ adalah **14**.

### Contoh 2 (Melibatkan Bilangan Negatif)
Diketahui matriks $C = \begin{pmatrix} 3 & -1 \\ 4 & 2 \end{pmatrix}$. Tentukan $\det(C)$!

**Penyelesaian:**
* $a = 3, b = -1, c = 4, d = 2$
$$\det(C) = (3 \cdot 2) - (-1 \cdot 4)$$
$$\det(C) = 6 - (-4)$$
$$\det(C) = 6 + 4$$
$$\det(C) = 10$$
Jadi, nilai determinan matriks $C$ adalah **10**.

---

## 4. Jenis Matriks Berdasarkan Nilai Determinan

Berdasarkan nilai determinannya, matriks persegi dibagi menjadi dua jenis:

1. **Matriks Singular**
   * Adalah matriks yang nilai determinannya **sama dengan nol** ($\det(A) = 0$).
   * Matriks singular **tidak memiliki invers** (kebalikan).
   * *Contoh:* $M = \begin{pmatrix} 6 & 3 \\ 4 & 2 \end{pmatrix} \rightarrow \det(M) = (6\cdot2) - (3\cdot4) = 12 - 12 = 0$.

2. **Matriks Non-Singular**
   * Adalah matriks yang nilai determinannya **tidak sama dengan nol** ($\det(A) \neq 0$).
   * Matriks non-singular **memiliki invers**.

---

## 5. Sifat-Sifat Determinan Matriks

Berikut adalah beberapa sifat penting determinan yang sering digunakan untuk mempermudah perhitungan:

1. **Sifat Transpose:** $\det(A^T) = \det(A)$
   *(Nilai determinan tidak berubah meskipun baris dan kolomnya ditukar).*
2. **Sifat Perkalian Matriks:** $\det(A \cdot B) = \det(A) \cdot \det(B)$
3. **Sifat Invers:** $\det(A^{-1}) = \frac{1}{\det(A)}$
4. **Sifat Perkalian Skalar:** $\det(k \cdot A) = k^2 \cdot \det(A)$ *(khusus untuk ordo $2 \times 2$, di mana $k$ adalah konstanta/skalar).*

---
*Tips Belajar: Berhati-hatilah dengan tanda negatif $(-)$ saat mengalikan elemen diagonal sekunder. Kesalahan umum yang sering terjadi adalah lupa mengubah tanda minus bertemu minus menjadi plus ($-\cdot- = +$).*
