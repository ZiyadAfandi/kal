# Tugas Aljabar Linear: Determinan dan Invers Matriks

## A. Perhitungan Determinan (Ekspansi Baris)

### 1. Matriks 2x2
$A = \begin{bmatrix} -7 & -5 \\ 1 & 4 \end{bmatrix}$

**Langkah-langkah:**
* Ekspansi baris ke-1:
* $\text{det}(A) = (-1)^{1+1}(-7)(4) + (-1)^{1+2}(-5)(1)$
* $\text{det}(A) = -28 + 5 = -23$

---

### 2. Matriks 3x3
$A = \begin{bmatrix} 0 & 2 & -3 \\ 1 & -2 & -1 \\ 0 & 0 & 1 \end{bmatrix}$

**Langkah-langkah:**
* Menggunakan ekspansi baris ke-3 (karena banyak angka nol):
* $\text{det}(A) = 0 + 0 + (-1)^{3+3}(1) \begin{vmatrix} 0 & 2 \\ 1 & -2 \end{vmatrix}$
* $\text{det}(A) = 1 \cdot (0 - 2) = -2$

---

### 3. Matriks 4x4
$A = \begin{bmatrix} 1 & -3 & 1 & 1 \\ -3 & 1 & 1 & 1 \\ 1 & 1 & -3 & 1 \\ 1 & 1 & 1 & -3 \end{bmatrix}$

**Hasil Akhir:**
* $\text{det}(A) = -32$

---

## B. Perhitungan Invers (Metode Adjoin)

### 4. Invers Matriks 2x2
$\text{det}(A) = -23$
$\text{adj } A = \begin{bmatrix} 4 & 5 \\ -1 & -7 \end{bmatrix}$

**Invers:**
$A^{-1} = \frac{1}{-23} \begin{bmatrix} 4 & 5 \\ -1 & -7 \end{bmatrix}$

---

### 5. Invers Matriks 3x3
$\text{det}(A) = -2$
$\text{adj } A = \begin{bmatrix} -2 & -2 & -8 \\ -1 & 0 & -3 \\ 0 & 0 & -2 \end{bmatrix}$

**Invers:**
$A^{-1} = \begin{bmatrix} 1 & 1 & 4 \\ 0.5 & 0 & 1.5 \\ 0 & 0 & 1 \end{bmatrix}$

---

### 6. Invers Matriks 4x4
$\text{det}(A) = -32$

**Invers:**
Setiap elemen matriks adjoin dibagi dengan -32, menghasilkan matriks dengan semua elemen bernilai $0.5$ atau $-0.5$ (tergantung pada perhitungan kofaktor spesifik tiap sel).
