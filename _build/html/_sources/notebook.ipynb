Kita akan memvisualisasikan perkalian matriks transformasi terhadap sebuah vektor.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan vektor v
v = np.array([1, 2])

# Matriks transformasi A (Rotasi & Skala)
theta = np.radians(45) # Rotasi 45 derajat
cos_t, sin_t = np.cos(theta), np.sin(theta)
A = np.array([
    [cos_t, -sin_t],
    [sin_t, cos_t]
]) * 1.5 # Perbesar skala 1.5x

# Transformasi vektor
Av = A.dot(v)
print(f"Vektor awal v: {v}")
print(f"Vektor setelah transformasi Av: {Av}")
```

## Visualisasi Vektor

Mari kita visualisasikan vektor awal dan vektor hasil transformasi menggunakan `matplotlib`.

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Plot vektor asal
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vektor Asli v')
# Plot vektor hasil transformasi
ax.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vektor Hasil Transformasi Av')

ax.set_xlim(-1, 4)
ax.set_ylim(-1, 4)
ax.grid(True, linestyle='--')
ax.axhline(0, color='black',linewidth=1)
ax.axvline(0, color='black',linewidth=1)
ax.set_aspect('equal')
ax.legend()
plt.title("Visualisasi Transformasi Vektor")
plt.show()
```
